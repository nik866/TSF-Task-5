```python
batting_first.head()
```

# Histogram Plotting

```python
plt.figure(figsize=(8,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel("runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
batting_first['winner'].value_counts().keys()
```

```python
batting_first['winner'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(8,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts())[0:5],color=["red","blue"])
plt.show()
```
